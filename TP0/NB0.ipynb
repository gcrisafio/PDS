```python
import numpy as np
import matplotlib.pyplot as plt

def plot_fasorial(vectors, etiquetas):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, polar=True)

    ax.set_facecolor('black')
    ax.tick_params(colors='white')
    ax.spines['polar'].set_color('black')

    for i, vector in enumerate(vectors):
        magnitude, angle, color = vector

        if i >= 6:
            magnitude *= 5  # Escala de corriente (para hacer los vectores de corriente ms largos)

        theta = np.radians(angle)
        width = np.radians(4)  # Ancho de los vectores en grados
        ax.bar(theta, magnitude, width=width, color=color, edgecolor='black', alpha=0.8)

        text_theta = theta + np.radians(7.5)  # Ajuste de posicin del texto
        ax.text(text_theta, magnitude, etiquetas[i], color=color, fontsize=12, ha='center', va='center')

    ax.set_yticklabels([])  # Desactivar etiquetas en el eje radial
    ax.set_xticklabels([])  # Desactivar etiquetas en el eje angular

    plt.show()


# Valores de ejemplo para los vectores
E = 220
Ua = (E, 0, 'c')
Ub = (E, -120, 'c')
Uc = (E, 120, 'c')
Uab = (E * np.sqrt(3), 30, 'b')
Ubc = (E * np.sqrt(3), -90, 'b')
Uca = (E * np.sqrt(3), 150, 'b')
Iab = (E * np.sqrt(3) / 20, -15, 'r')
Ibc = (E * np.sqrt(3) / 20, -135, 'r')
Ica = (E * np.sqrt(3) / 20, 105, 'r')
Ia = (np.sqrt(3) * E * np.sqrt(3) / 20, -45, 'm')
Ib = (np.sqrt(3) * E * np.sqrt(3) / 20, -165, 'm')
Ic = (np.sqrt(3) * E * np.sqrt(3) / 20, 75, 'm')

vectores =  [ Ua ,  Ub ,  Uc ,  Uab ,  Ubc ,  Uca ,  Iab ,  Ibc ,  Ica ,  Ia ,  Ib ,  Ic ]
etiquetas = ["Ua", "Ub", "Uc", "Uab", "Ubc", "Uca", "Iab", "Ibc", "Ica", "Ia", "Ib", "Ic"]
plot_fasorial(vectores, etiquetas)
```
