```python
# Importamos la librería numpy para usar funciones matemáticas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la función senoidal con los parámetros que nos piden
def mi_funcion_sen(vmax, dc, ff, ph, nn, fs):
  
  ## Tiempo
  # Creamos un vector de tiempo con nn muestras y fs frecuencia de muestreo
  Ts = 1/fs # step = periodo de muestreo Ts (ADC) 
  T_exploracion = nn*Ts # Tiempo de exploracion = N . Ts
  tt = np.arange(start = 0, stop = T_exploracion, step = Ts)

  ## Signal
  # Calculamos la señal senoidal con la fórmula: x(t) = vmax * sin(2 * pi * ff * t + ph) + dc
  xx = vmax * np.sin(2 * np.pi * ff * tt + ph) + dc
  # Devolvemos el vector de tiempo y la señal senoidal
  return tt, xx

fs = 10. #[Hz] Definido por ADC 
  # NOTA: usar escalares con X. para usar variables tipo enteras en procesador 
N = fs # muestras (Normalizamos para ver mejor el grafico)

# In: 
#   .la amplitud máxima de la senoidal (volts)
#   .su valor medio (volts)
#   .la frecuencia (Hz)
#   .la fase (radianes)
#   .la cantidad de muestras digitalizada por el ADC (N muestras)
#   .la frecuencia de muestreo del ADC.


# Grafico
#plt.figure(figsize=(10, 6)) # Crear una figura de 10 x 6 pulgadas
tt, xx = mi_funcion_sen( vmax = 1, dc = 0, ff = 1, ph=0, nn = N, fs = fs)
plt.plot(tt, xx);

# Configurar el estilo del gráfico
#plt.style.use('dark_background')
# plt.grid(color='white', linestyle='--', linewidth=0.25)
```

```python
# Implementar alguna otra señal propia de un generador de señales. 

import matplotlib.pyplot as plt # Importar la librería matplotlib
import numpy as np # Importar la librería numpy
from scipy.signal import square # Importar la función square

def mi_funcion_cuad(vmax, dc, ff, ph, nn, fs):
  
  ## Tiempo
  # Creamos un vector de tiempo con nn muestras y fs frecuencia de muestreo
  Ts = 1/fs # step = periodo de muestreo Ts (ADC) 
  T_exploracion = nn*Ts # Tiempo de exploracion = N . Ts
  tt = np.arange(start = 0, stop = T_exploracion, step = Ts)

  ## Signal
  # Calculamos la señal cuadrada con la fórmula: x(t) = vmax * square(2 * pi * ff * t + ph, duty) + dc
  # El parámetro duty es el ciclo de trabajo de la señal cuadrada, que debe estar entre 0 y 1
  duty = 0.5 # Por ejemplo, ponemos un ciclo de trabajo de 0.5, que significa que la señal es positiva el 50% del tiempo
  xx = vmax * square(2 * np.pi * ff * tt + ph, duty) + dc
  # Devolvemos el vector de tiempo y la señal cuadrada
  return tt, xx


fs = 1000. #[Hz] Definido por ADC 
  # NOTA: usar escalares con X. para usar variables tipo enteras en procesador 
N = fs # muestras (Normalizamos para ver mejor el grafico)

tt, xx = mi_funcion_cuad( vmax = 1, dc = 0, ff = 5, ph=0, nn = N, fs = fs)
plt.figure(figsize=(10, 6)) # Crear una figura de 10 x 6 pulgadas
plt.plot(tt, xx);

# Configurar el estilo del gráfico
plt.style.use('dark_background')
plt.grid(color='white', linestyle='--', linewidth=0.25)
```
